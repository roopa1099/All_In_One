```python
from ucimlrepo import fetch_ucirepo
# https://archive.ics.uci.edu/dataset/2/adult   --- uci id=2  => Prediction task is to determine whether a person's income is over $50,000 a year.
# https://archive.ics.uci.edu/dataset/336/chronic+kidney+disease   ---- ckd(id  = 336)
```

```python
census = fetch_ucirepo(id=2)
```

```python
# data (as pandas dataframes) 
X = census.data.features 
y = census.data.targets 


# This variable represents your Independent Variables (or features). It is a collection of all the data points the model will use to learn patterns and make predictions.

#     Format: Usually a pandas DataFrame.

#     Content: In a census dataset, this includes columns like age, education, occupation, hours-per-week, and marital-status.

#     Purpose: These are the "clues" the model looks at to understand the underlying trends.

# 2. y = census.data.targets

# This variable represents your Dependent Variable (or target/label). This is the specific thing you want the model to predict.

#     Format: Usually a pandas Series or a single-column DataFrame.

#     Content: In the famous Census Income dataset, this is typically a column indicating whether a person earns >50K or <=50K per year.

#     Purpose: This is the "ground truth" used to train the model and later evaluate if its guesses are correct.


print("X -------> \n", X)
print("------------------------NEXT---------------------")
print("Y -------> \n", y)
# metadata 
print(census.metadata) 
  
# variable information 
print(census.variables) 
```

```python
# What to do if missing values exist:
#  1. Ignore : Drop the rows (means check total rows having different null values)
#                 - rows with missing data / total rows << 1
#                 - missing values are randomly distributed (means there is no pattern in missing values for example: 
#                    if age is missing for mainly female employees, we cannot ignore as it has a pattern), correlation of missing values

total_rows = len(X)
print(total_rows) 

workclass_null = X["workclass"].isnull().sum()
print("Workclass Null : ",workclass_null) 

print(workclass_null/total_rows,"\n")

occupation_null = X["occupation"].isnull().sum()
print("Occupation Null : ",occupation_null) 

print(occupation_null/total_rows,"\n")

native_country_null = X["native-country"].isnull().sum()
print("native_country Null : ",native_country_null) 

print(native_country_null/total_rows)
```

```python

#  if while reading we flag ? as nan it will read that also as null ***** how to do it.


import numpy as np

df = census.data.features
df = df.replace("?", np.nan)

df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
# cbar = color bar (the scale on the side)
# cmap="viridis", cmap = color map
sns.heatmap(df.isna(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()

#  Workclass and occupation missing looks corrrelated
```

```python
# 2. Imputation : Filling missing values with median: Not for categorical
                 # mean preferred if data is normally distributed, not outliers much as mean is affected by outliers

 # 3. For categorical: Plot hostigram to understand how uniq values are distributed in null value column
```

```python
import matplotlib.pyplot as plt

#  for categorical we plot bar, for numerical we take hist
#  plt.hist(df["workclass"], bins=20)
df["workclass"].value_counts().plot(
    kind="bar",
    figsize=(8, 4)
)
plt.xlabel("Workclass")
plt.ylabel("Count")
plt.title("Distribution of Workclass")
plt.show()
# skewed to private
```

```python
df["occupation"].value_counts().plot(
    kind="bar",
    figsize=(8, 4)
)
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.title("Distribution of Occupation")
plt.show()
```

```python
df["native-country"].value_counts().plot(
    kind="bar",
    figsize=(8, 4)
)
plt.xlabel("native-country")
plt.ylabel("Count")
plt.title("Distribution of Native-country")
plt.show()
# Skewed to US
```
