```python
test = np.array([1,1.2,4])
# NumPy always promotes data to the "most complex" type present in the input.An Integer can be turned into a Float without 
# losing value ($1$ becomes $1.0$).
# A Float cannot be turned into an Integer without losing the decimal ($1.2$ would become $1$, losing the $.2$).

test.dtype
```

```python
test = np.array([True,True])
test.dtype
```

```python
# the Boolean type is actually a "sub-type" of integers:

# True is treated as 1.
# False is treated as 0.

test = np.array([True,12])
test.dtype
```

```python
# the Boolean type is actually a "sub-type" of integers:

# True is treated as 1.
# False is treated as 0.


test = np.array([True,12.2])
test.dtype
```

```python
# ******IMPORTANT*********
# Int and Float => type changes to float
# Int and boolean => type in int as boolean is also considered int
# Int and string => defaults to U21
# String => U(max length of string)
# Float and boolean => float
# Float and string => U32
# bool and string => 4/5 letter string, n letter string => U(max(n,4)


# A 64-bit float can have many decimal places, an "e" for scientific notation (like 1.2e+10), and a decimal point.

# To ensure no precision is lost when a float is turned into text, NumPy allocates a larger default buffer—usually 32 characters.

test = np.array(["fdf",12.2])
test.dtype
```

```python

# Further features
test_array = [3,6,2,8,34,65,13,98]
test_numpy = np.array(test_array)

type(test_numpy)
```

```python
test_array>=18
```

```python
# You can just write a expression using numpy array's name and it does the operation on every element of array

test_numpy >= 18
```

```python
# To get specific index of elements
test_numpy[[3,1,4]] 
test_numpy[[2]]

# for 1D array you cannot have mutiple array to get index 
# test_numpy[[3,1,4],[[2,1,0]]] 
```

```python
# An array must be a perfect rectangle. In your code, the first row has 7 numbers, but the second row only has 5 numbers.
test_numpy_2D = np.array([[4,2,7,45,13,68,98],[34,21,67,87,77]])
```

```python
# Fancy indexing
test_numpy_2D = np.array([[4,2,7,45,98],[34,21,67,87,77]])
# below if you see we can pickup elements using fancy indexing, but need to remember, first array is row seconf is column
# (1,2) (1,1) (0,0)
test_numpy_2D[[1,1,0],[2,1,0]] 
```

```python
# np.where always returns a Tuple,
# np.where returns index
np.where(test_numpy%2 == 0)
```

```python
# In a 1D array, you only need one list of indices to find the values.

# In a 2D array, you need two lists (one for Rows, one for Columns).

# In a 3D array, you need three lists.
np.where(test_numpy_2D%2==0)
#  responser is one for rows, one for column
```

```python
test_numpy[4]
```

```python
test_numpy_2D[1][2]
```

```python
# Fancy indexing follows the shape of the index list you provide.
arr = np.array([10, 20, 30, 40])
indices = np.array([[0, 1], 
                    [2, 3]])

print(arr[indices])

```
