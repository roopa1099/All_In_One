```python
"""
Grouping & Aggregation
Q17

Avg salary per department

Max rating per location

Multiple aggregations using agg()
"""

emp = pd.read_csv("employees.csv")

print(emp.groupby("dept")["salary"].mean()) # Max rating per location
print("\n")

print(emp.groupby("location")["rating"].max())
print("\n")

print(emp.groupby("dept").agg({"salary":["mean","median"], "rating":"mean"})) 
print("\n")
```

```python
"""
Q18

Group by department and calculate:

mean salary

total experience

employee count
"""


print(emp.groupby("dept").agg({"salary":"mean","experience":"sum","emp_id":"count"}))
print("\n")
```

```python
"""
Transform vs Aggregate
Q19

Subtract department mean salary from each employee

Add column salary_vs_dept_avg
"""

emp["salary_avg_sub"] = emp["salary"] - emp.groupby("dept")["salary"].transform("mean") # imp****

# transform() broadcasts the correct department mean to every row in that department, so subtraction happens correctly row-by-row.
print(emp)
print("\n")



emp["salary_vs_dept_avg"]
```
